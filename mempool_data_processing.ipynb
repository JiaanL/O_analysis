# Maker

```python
from_block = 12466826
df_list = []
while True:
    if from_block == 14957468: break
    to_block = from_block + 118602
    data_path = f"data/crawled/maker_dao_eth_usd_median_gas_info_{from_block}_{to_block}.csv"
    df_list.append(pd.read_csv(data_path))
    from_block = to_block
log_tx_hash = pd.concat(df_list).reset_index(drop=True)

from_block = 12466826
df_list = []
while True:
    if from_block == 14957468: break
    to_block = from_block + 118602
    data_path = f"data/crawled/maker_dao_eth_usd_median_{from_block}_{to_block}.csv"
    df = pd.read_csv(data_path)

    cut_gap = int(len(df['Data'].iloc[0])/2)
    cut_start = 0
    for i, new_col in enumerate("price age".split(" ")):
        cut_end = cut_start + cut_gap
        df[new_col] = df['Data'].apply(lambda x: x[cut_start:cut_end])
        cut_start = cut_end
    del df['Data']

    def reformat_log_maker_dao_median(raw_data):
        data = copy.deepcopy(raw_data)
        for col_name in 'price age'.split(' '):
            data[col_name] = raw_data[col_name].map(hex2int)
        return data
    df = reformat_log_maker_dao_median(df)
    df.price = df.price * 10e-19

    df.sort_values(['blockNum', 'index'])
    df_list.append(df)

    from_block = to_block
log_price = pd.concat(df_list).reset_index(drop=True)

df_list = []
for i in range(7):
    data_path = f"data/crawled/blockTime_data_{i}.csv"
    df_list.append(pd.read_csv(data_path))
blockTime_df = pd.concat(df_list).reset_index(drop=True)
# blockTime_df
```

```python
maker_mempool_df  = log_tx_hash.merge(log_price, how='left', on=['blockNum', "index"])#  pd.concat([log_tx_hash, log_price], axis=1).iloc[:,:-2]
maker_mempool_df  = pd.merge(maker_mempool_df, blockTime_df, on="blockNum", how="left")
maker_mempool_df = maker_mempool_df['blockNum index blockTime txHash price'.split(' ')]
# maker_mempool_df['current'] = maker_mempool_df['current'].apply(lambda x: hex2int(x)*10e-9)
maker_mempool_df = maker_mempool_df.dropna()
maker_mempool_df
```

```python
maker_mempool_df.to_csv("data/maker_mempool_df.csv", index=False)
```
