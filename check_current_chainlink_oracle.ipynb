```python
def hex2int(x):
    return int(x, 16)

def reformat_log(raw_data):
    data = copy.deepcopy(raw_data)
    for col_name in 'topic0 current roundId updatedAt'.split(' '):
        data[col_name] = raw_data[col_name].map(hex2int)
    data['updatedAt'] = pd.to_datetime(data['updatedAt'],unit='s')
    data['current'] = data['current'] / 10**8
    return data

def read_csv_files(file_name, blockFrom = 12466826, blockGap = 118602, blockMax = 14838875, print_file_name=False):
    df_list = []
    while blockFrom <= blockMax:
        blockTo = blockFrom + blockGap
        filename = f'data/crawled/{file_name}_{blockFrom}_{blockTo}.csv'
        if print_file_name: print(f'read {filename}')
        df = pd.read_csv(filename, index_col=False)
        blockFrom = blockTo
        df_list.append(df)
    return pd.concat(df_list)
```

```python
df = read_csv_files('aave_liquidationcall')
# only consider eth as collateral
df['collateralAsset'] = df['topic1']
df['debtAsset'] = df['topic2']
df['users'] = df['topic3']
del df['topic0'], df['topic1'], df['topic2'], df['topic3']

cut_gap = int(len(df['Data'].iloc[0])/4)
cut_start = 0
for i, new_col in enumerate("dabtToCover liquidatedCollateralAmount liquidator receiveAToken".split(" ")):
    cut_end = cut_start + cut_gap
    df[new_col] = df['Data'].apply(lambda x: x[cut_start:cut_end])
    cut_start = cut_end
del df['Data']
df.sort_values(['blockNum', 'index'])

aave_liquidation_df = df
# only need eth - usdc
aave_liquidation_df = aave_liquidation_df[aave_liquidation_df['collateralAsset'] == '0x000000000000000000000000c02aaa39b223fe8d0a0e5c4f27ead9083c756cc2']
aave_liquidation_df = aave_liquidation_df[aave_liquidation_df['debtAsset'] == '0x000000000000000000000000a0b86991c6218b36c1d19d4a2e9eb0ce3606eb48']

aave_liquidation_df['dabtToCover'] = aave_liquidation_df['dabtToCover'].apply(hex2int)
aave_liquidation_df['liquidatedCollateralAmount'] = aave_liquidation_df['liquidatedCollateralAmount'].apply(hex2int)
aave_liquidation_df['receiveAToken'] = aave_liquidation_df['receiveAToken'].apply(lambda x: bool(int(hex2int(x))))
aave_liquidation_df = aave_liquidation_df.reset_index(drop=True)
aave_liquidation_df = aave_liquidation_df['blockNum index dabtToCover liquidatedCollateralAmount'.split(' ')]
aave_liquidation_df
```
